```python
import numpy as np
import matplotlib.pyplot as plt
import nibabel as nib
from data_manager import Dataset_2D, df_to_Dataset, build_data
import pandas as pd
```

```python
data_csv = '/home/ge.polymtl.ca/p120530/Training/Kaggle/train.csv'
base_dir = '/home/ge.polymtl.ca/p120530/Training/Kaggle/train_nifti'



# Build the data
pd_train_data, pd_val_data, num_classes = build_data(data_csv, base_dir)

# Create the training and validation datasets
train_dataset = df_to_Dataset(pd_train_data, val=True)
val_dataset = df_to_Dataset(pd_val_data, val=True)

shapeX = []
shapeZ = []
shapeY = []

# go trough the dataset, plot the shape of the images
for i in range(len(train_dataset)):
    img, mask = train_dataset[i]
    shapeX.append(img.shape[2])
    shapeY.append(img.shape[3])
    shapeZ.append(img.shape[4])
for i in range(len(val_dataset)):
    img, mask = val_dataset[i]
    shapeX.append(img.shape[2])
    shapeY.append(img.shape[3])
    shapeZ.append(img.shape[4])
```

```python
# Scatter the shape of the images
plt.scatter(shapeX, shapeZ,
             c='r', marker='o', label='Image', alpha=0.2)
plt.xlabel('Shape[0]')
plt.ylabel('Shape[2]')
plt.title('Image shape')
plt.legend()
plt.show()
```

```python
# Scatter the shape of the images
plt.scatter(shapeX, shapeY,
             c='r', marker='o', label='Image', alpha=0.2)
plt.xlabel('Shape[0]')
plt.ylabel('Shape[1]')
plt.title('Image and Mask shape')
plt.legend()
plt.show()
```
