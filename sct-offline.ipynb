```python
# Path to the directory where you want to change permissions
bin_folder = '/kaggle/working/spinalcordtoolbox/bin/'

# Traverse the directory and its subdirectories
for root, dirs, files in os.walk(bin_folder):
    # Change permissions for each directory
    for dir_name in dirs:
        dir_path = os.path.join(root, dir_name)
        os.chmod(dir_path, stat.S_IRWXU)  # Grant read, write, execute to user

    # Change permissions for each file
    for file_name in files:
        file_path = os.path.join(root, file_name)
        os.chmod(file_path, stat.S_IRWXU)  # Grant read, write, execute to user
```
